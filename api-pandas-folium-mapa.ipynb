Para señalar las 20 coordenadas de la provincia de Zaragoza, realizamos en este notebook el mismo procedimiento que en el documento llamado api-pandas-folium.ipynb desde cero en una misma operación. Finalmente, con la orden `mapa`, se imprimieron todos los marcadores solicitados. 

```python
!pip install pandas folium
import pandas as pd
import folium
url = 'https://www.zaragoza.es/sede/servicio/transporte\
/accidentalidad-trafico/accidente.csv?rows=20'
coord = [41.64,-0.88]
mapa = folium.Map(location=coord)
df = pd.read_csv(url,delimiter=';')
longitudes = []
for i in df['geometry']:
    longlat = i.split(',')
    longitudes += [float(longlat [0])]
latitudes = []
for i in df['geometry']:
    longlat = i.split(',')
    latitudes += [float(longlat [1])]
df_coord = pd.DataFrame({'long':longitudes,\
                         'lat':latitudes})
df_tipo = pd.concat([df['type'], df_coord], axis=1)
for index, fila in df_tipo.iterrows():
    marcador = folium.Marker([fila['lat'], fila['long']],\
    popup=fila['type'])
    mapa.add_child(marcador)
mapa
```

## Mapa con icono y guardar mapa

Para concluir la **Actividad dirigida 4**, se realizó otro mapa pero está vez de Madrid, añadiendo los datos `latitud` y `longitud`. Después, con la función `mapa = folium.Map(location=coord, tiles='Stamen Terrain')`, se manipularon los datos para poder visualizar y personalizar el mapa. 

Con respecto a la función `folium.Marker`, esta se utilizó para crear un punto en el mapa, el cual fue cambiado al color verde con la orden `icon=folium.Icon(color='green')`. Finalmente, con la función `mapa.add_child`, se asignaron al mapa los cambios introducidos al marcador. 

La orden `mapa` imprimió el gráfico con estas especificaciones.

```python
coord = [40.4430801,-3.7130593]
mapa = folium.Map(location=coord, tiles='Stamen Terrain')
marcador = folium.Marker(coord, icon=folium.Icon(color='green'))
mapa = mapa.add_child(marcador)
mapa
```

Para guardar el mapa como html hemos utilizado la función `mapa.save('mapa.html')`

```python
mapa.save('mapa.htlm')
```
